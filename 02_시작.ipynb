```python
import numpy as np
print(np.__version__)
my_arr=np.array([[10,20,30],[40,50,60]])
print(my_arr)
print(type(my_arr))
print(my_arr[0][2])
print(np.sum(my_arr)) #np 내장함수 (for 대신 작성)
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4,5],[1,4,9,16,25]) # x축과 y축의 값을 python의 list로 전달
plt.show()
```

```python
#산점도
plt.scatter([1,2,3,4,5],[1,4,9,16,25])
plt.show() #안써도 그래프는 그려주지만, 더 깔끔하게 그래프 표현 가능
```

```python
#난수 생성과 넘파이 배열로 산점도 그리기
x=np.random.randn(1000) #표준 정규 분포를 따르는 1000 개의 난수 생성
y=np.random.randn(1000)
plt.scatter(x,y)
plt.show()
```
