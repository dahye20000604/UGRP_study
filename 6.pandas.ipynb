## NaN값 처리 

```python
#NaN값 확인
train_data.info()
```

```python
#NaN 데이터 삭제
train_data.dropna() #nan이 하나라도 있으면 sample 누락
```

```python
train_data.dropna(subset=['Age','Cabin'])
```

```python
#column 자체지우기
train_data.dropna(axis=1) 
```

```python
#NaN값 대체 평균 or 생존자/사망자 별 평균
train_data['Age'].fillna(train_data['Age'].mean()) #inplac=true 주면 data 바로 바뀐다.
```

```python
#생존자 나이 평균, 사망자 나이 평균
m1=train_data[train_data['Survived']==1]['Age'].mean()
m2=train_data[train_data['Survived']==0]['Age'].mean()
m1
```

```python
train_data[train_data['Survived']==1]['Age'].fillna(m1)
```

```python
train_data[train_data['Survived']==0]['Age'].fillna(m2)
```

```python
train_data.loc[train_data['Survived']==1,'Age']=train_data[train_data['Survived']==1]['Age'].fillna(m1) #survived==1인 행의 age에 대입
train_data.loc[train_data['Survived']==0,'Age']=train_data[train_data['Survived']==0]['Age'].fillna(m2)
train_data
```

## 숫자데이터와 범주형 데이터

```python
train_data.info()
```

```python
#숫자가 카테고리가 나누어주거나 구분주어지는건 범주형데이터 
# 범주형 데이터는 버리거나 숫자형으로 변환
train_data=pd.read_csv('./train.csv')
train_data.head()
```

```python
train_data['Pclass']=train_data['Pclass'].astype(str)
train_data
```

```python
import math
def age_categorize(age):
    if math.isnan(age):
        return -1
    return math.floor(age/10)*10
    
```

```python
#apply: for문 안쓸 수 있다.
train_data['Age'].apply(age_categorize)
```

```python
#실제로는 숫자형 데이터만 사용!(연산가능)
pd.get_dummies(train_data,columns=['Pclass','Sex','Embarked'],drop_first=False) #onehotencoding
```

```python
pd.get_dummies(train_data,columns=['Pclass','Sex','Embarked'],drop_first=True) #변수를 하나라도 줄이기 위해, 나머지 하나의 범주형 데이터는 유추가능
```
