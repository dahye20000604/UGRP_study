#### requests 모듈
1. http request/response를 위한 모듈
2. HTTP method를 메소드 명으로 사용하여 request 요청 예) get, post

```python
import requests
```

#### get 요청하기
1. http get 요청하기
2. query parameter 이용하여 데이터 전달하기

```python
url = 'https://news.v.daum.net/v/20190728165812603'
resp=requests.get(url) #이 주소에 해당하는 서버에 get요청해서 응답을 가져오자

resp #요청 성공 = 200
resp.text #페이지 소스보기랑 동일
```

#### post 요청하기
1. http post 요청하기
2. post data 이용하여 데이터 전달하기

```python
#보통 로그인에 사용
url='https://www.kangcom.com/member/member_check.asp'
data={
    'id':'macmath22',
    'pwd':'Test1357!'
}
resp=requests.post(url,data=data)
resp.text
```

#### HTTP header 데이터 이용하기
1. header 데이터 구성하기
2. header 데이터 전달하기

```python
url = 'https://news.v.daum.net/v/20190728165812603'
headers={'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.141 Safari/537.36'}
resp=requests.get(url, headers=headers)
resp.text
```
