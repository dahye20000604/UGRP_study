### 02. 조건문(if, elif, else) 활용하기

```python
a = 10
if a % 2 ==0:
    print(a/2)
#if와 else 사이에는 어떤 코드도 들어갈 수 없다.
else:
    print((a+1)/2)
```

```python
a = 18
if a % 4 == 0:
    print('a is divisible by 4')
elif a % 4 == 1:        #elif 전에는 항상 if가 있어야함.  
    print('a % 4 is 1')
elif a % 4 == 2:
    print('a % 4 is 2')
else:                   #else는 조건을 명시할 수 없다.
    print('a % 4 is 3')
```

```python
a = 16
if a % 4 == 0:
    print('a is divisible by 4')
if a % 4 == 1:        
    print('a % 4 is 1')
if a % 4 == 2:
    print('a % 4 is 2')
else:                   
    print('a % 4 is 3')
```

```python
#중첩 조건문(nested condition)

a = 10
b = 9
c = 8

if a == 10:
    if c == 8:
        if b == 8:
            print('a is ten and b is 8')
        else:
            print('a is ten and b is not 8')
```

### 03. 반복문 이해하기(while) - 1

```python
a=[1, 10,9,24]

i = 0
while i < len(a):
    print('value :',a[i], ', index :',i)
    i += 1
print('hahaha')
```

```python
a = [1, 10, 9, 24, 25, 26]

i=0
while i < len(a):
    if a[i] > 20:
        print(a[i])
    i += 1
    
```

```python
a = [1, 10, 9, 24, 25, 26]

i=0
while i < len(a):
    if a[i] % 2: #홀수 일때
        print(a[i])
    else:
        print(a[i] / 2)
    i += 1
    
```
