### 02. 조건문(if, elif, else) 활용하기

```python
a = 10
if a % 2 ==0:
    print(a/2)
#if와 else 사이에는 어떤 코드도 들어갈 수 없다.
else:
    print((a+1)/2)
```

```python
a = 18
if a % 4 == 0:
    print('a is divisible by 4')
elif a % 4 == 1:        #elif 전에는 항상 if가 있어야함.  
    print('a % 4 is 1')
elif a % 4 == 2:
    print('a % 4 is 2')
else:                   #else는 조건을 명시할 수 없다.
    print('a % 4 is 3')
```

```python
a = 16
if a % 4 == 0:
    print('a is divisible by 4')
if a % 4 == 1:        
    print('a % 4 is 1')
if a % 4 == 2:
    print('a % 4 is 2')
else:                   
    print('a % 4 is 3')
```

```python
#중첩 조건문(nested condition)

a = 10
b = 9
c = 8

if a == 10:
    if c == 8:
        if b == 8:
            print('a is ten and b is 8')
        else:
            print('a is ten and b is not 8')
```

### 05. 반복문 이해하기(for), 연습문제 - 1

```python
a = [1,2,4,3,5]
for i in a:
    print(i, i*2)
print('hahaha')
```

```python
#문자열의 아이템 출력하기
a = 'hello world'
for character in a:
    print(character)
```

```python
a = [1, 10, 3, 4, 5]
for num in a:
    if num % 2 ==0:
        print(num/2)
    else:
        print(num+1)
```

```python
#dict의 아이템 출력하기
a = {'korea':'seoul', 'japan' : 'tokyo', 'canada': 'ottawa'}
for key in a:
    print(key, a[key])
for k in a.values():
    print(k)
for key,value in a.items():
    print(key, value)
```

```python
#index가 필요한 경우
a = [1,2,3,4,5]
for index, num in enumerate(a):
    if index > 3:
        print(index, num)
```

```python
#break
a = [100, 90, 80, 70, 60, 50]
for num in a:
    if num < 80:
        break
    print(num)
```

```python
#continue
a = [100, 90, 80, 70, 60, 50]
for num in a:
    if num >= 60 and num <= 70:
        continue
    print(num)
```

```python
#loop 중첩
a = [1,2,4]
for i in a:
    for j in a:
        print(i*j)
```

### 07. 조건문, 반복문 연습 문제 풀이

```python
#1. 구구단을 2-9단까지 출력
for i in range(2,10):
    for j in range(1,10):
        print(i*j,end=' ')
    print()

print()

x=2
while x<10:
    for i in range(1,10):
        print(x*i,end=' ')
    print()
    x+=1
```

```python
#2. 1-100까지 정수 중 2의 배수 또는 11의 배수를 모두 출력
for i in range(1,101):
    if i%2==0 or i%11==0:
        print(i, end=' ')
```

```python
#3. a = [22,1,3,4,7,98,21,55,87,99,19,20,45]에서 최대값과 최소값을 찾으시오. (sorted, sort사용금지)
a = [22,1,3,4,7,98,21,55,87,99,19,20,45]
'''a.sort()
a[0]
a[-1]
'''
max=a[0]
min=a[0]
for i in a[1:]:
    if max<i:
        max=i
    elif min>i:
        min=i
print(max,min)
```

```python
#4. a = [22,1,3,4,7,98,21,55,87,99,19,20,45]에서 평균 구하기
a = [22,1,3,4,7,98,21,55,87,99,19,20,45]
sum=0
for i in a:
    sum+=i
print(sum/len(a))
```
